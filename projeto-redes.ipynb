# Trabalho de Redes Sociais

## Descrição
Analisando a relação entre a centralidade de um jogador e sua perfomance em embates.

## Dados
Os dados foram retirados do [Kaggle](https://www.kaggle.com/pschale/mlb-pitch-data-20152018).

## Importando as bibliotecas

```
# Insert code here.
import numpy as np
import pandas as pd
pd.set_option('mode.chained_assignment', None)
import freeman
```

## Carregando os dados.

```
at_bats = pd.read_csv('atbats.csv')
games = pd.read_csv('games.csv')
players = list(at_bats.pitcher_id.unique()) + list(at_bats.batter_id.unique())
```

## Criando um CSV com os times de cada embate

```
def get_pitcher_team(row, games):
    if row.top:
        team = "home_team"
    else:
        team = "away_team"
    return games[games.g_id == row.g_id][team].values[0]

def get_batter_team(row, games):
    if not row.top:
        team = "home_team"
    else:
        team = "away_team"
    return games[games.g_id == row.g_id][team].values[0]
```

## _!!!! SÓ RODE ESSA CÉLULA CASO TENHA FEITO ALGUMA ALTERAÇÃO !!!!_

```
data = {}
at_bats_temp = at_bats[:]
pitcher_team = []
batter_team = []
for index, row in at_bats.iterrows():
    pitcher_team.append(get_pitcher_team(row, games))
    batter_team.append(get_batter_team(row, games))
        
```

```
at_bats.loc[:, "pitcher_team"] = pitcher_team
at_bats.loc[:, "batter_team"]  = batter_team
```

## Criando a composição dos times a cada jogo

```
data = {}
for index, row in at_bats.iterrows():
    game = row['g_id']
    game_dict = data.get(game, False)
    if game_dict:
        data[game][row['pitcher_team']].append(row['pitcher_id'])
        data[game][row['batter_team']].append(row['batter_id'])
    else:
        data[game] = {
            row['pitcher_team']: [row['pitcher_id']],
            row['batter_team']: [row['batter_id']]
        }
for game, teams in data.items():
    for team, players in teams.items():
        data[game][team] = list(set(players))
```

```
data
```

```
connections = {}
for game in data.values():
    for team, players in game.items():
        for index in range(len(players) - 2):
            player = players[index]
            player_connections = connections.get(player, None)
            if not player_connections:
                connections[player] = set()
            for i in range(index + 1, len(players)):
                partner = players[i]
                connection = connections.get(partner, None)
                if not connection or player not in connection:
                    connections[player].add(partner)
for player in players:
    if player not in connections:
        connections[player] = []
```

```
connections
```

```
gml = 'graph [\n\tdirected 0\n\n'
```

```
gml = 'graph [\n\tdirected 0\n'
for player in connections.keys():
    gml += f"\n\tnode [\n\t\tid {player}\n\t]"
gml += '\n'
for player, player_connections in connections.items():
    for connection in player_connections:
        gml += f'\n\tedge [\n\t\tsource {player}\n\t\ttarget {connection}\n\t]'
gml += '\n]'
with open('network.gml', 'w') as f:
    f.write(gml)
```
